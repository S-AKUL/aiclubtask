```python
firstquartile=power_df['demand_mw'].quantile(0.25)
thirdquartile=power_df['demand_mw'].quantile(0.75)
IQR =thirdquartile-firstquartile
```

```python
lower=firstquartile-1.5*IQR
upper=thirdquartile+1.5*IQR
print(lower,upper)
```

```python
power_df['demand_mw']=power_df['demand_mw'].clip(lower,upper)
```

```python
plt.figure(figsize=(15,5))
plt.plot(power_df['datetime'],power_df['demand_mw'])
plt.show()
```

```python

rolling_median=power_df['demand_mw'].rolling(window=24,center=True).median() #creates a series of median
difference=abs(power_df['demand_mw']-rolling_median)
threshold=difference.std()*3
outliers=difference>threshold #outliers is binary mask (series)
power_df.loc[outliers,'demand_mw']=rolling_median[outliers]  # masking the table and assigning it its respective 24 hr window median value
power_df['demand_mw']=power_df['demand_mw'].interpolate()
```

```python
plt.figure(figsize=(15,5))
plt.plot(power_df['datetime'],power_df['demand_mw'])
plt.show()
```

```python
#  removes the lower outliers
lower_threshold=rolling_median *0.4
low_outliers=power_df['demand_mw']<lower_threshold
power_df.loc[low_outliers,'demand_mw']=rolling_median[low_outliers]
power_df['demand_mw']=power_df['demand_mw'].interpolate()
```

```python
plt.figure(figsize=(15,5))
plt.plot(power_df['datetime'], power_df['demand_mw'])
plt.show()
```

```python
y_test=test_df['target']
```

```python
print("x_train:", x_train.shape)
print("y_train:", y_train.shape)

print("x_test:", x_test.shape)
print("y_test:", y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mae = mean_absolute_error(y_test, y_pred)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
print("MAE:", mae)
print("RMSE:", rmse)
```

```python
plt.figure(figsize=(15,5))
plt.plot(y_test.values[:500], label='Actual')
plt.plot(y_pred[:500], label='Predicted')
plt.legend()
plt.title("actual vs predicted ")
plt.show()
```

```python
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
print("MAPE:", mape)
```

```python
from xgboost import XGBRegressor
```

```python
model_xgb = XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    n_jobs=-1
)
```

```python
x_train.dtypes
```

```python
model_xgb.fit(x_train, y_train)
```

```python
y_pred_xgb = model_xgb.predict(x_test)
```

```python
mae_xgb = mean_absolute_error(y_test, y_pred_xgb)
rmse_xgb = np.sqrt(mean_squared_error(y_test, y_pred_xgb))
mape_xgb = np.mean(np.abs((y_test - y_pred_xgb) / y_test)) * 100

print("MAE (XGB):", mae_xgb)
print("RMSE (XGB):", rmse_xgb)
print("MAPE (XGB):", mape_xgb)
```

```python
importance = pd.Series(model_xgb.feature_importances_, index=x_train.columns)
importance.sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(15,5))
plt.plot(y_test.values[:1000], label='Actual')
plt.plot(y_pred_xgb[:1000], label='XGBoost')
plt.legend()
plt.title(" Actual vs Predicted using xgb")
plt.show()
```
